```python
import plotly
import plotly.offline as py
```

```python
from plotly.offline import plot, iplot
plotly.offline.init_notebook_mode(connected= True)
import plotly.graph_objs as go
```

```python
pmtros = {
    'PREDICT_RANGE_def': 150,
    'START_DATE_def': "1/22/20",
    'COUNTRY_CSV_def': ['Brazil'],
    'csv_confirmed': 'tables/casosXdia.csv',
    'csv_confirmedAc': 'tables/casosConfirmadosAc.csv',
    'csv_death': 'tables/obitosXdia.csv',
    'csv_deathAc': 'tables/numeroObitosAc.csv'
    
}

lsTables = ['csv_confirmed','csv_confirmedAc','csv_death','csv_recovered']
```

```python
dfConf = pd.read_csv(pmtros['csv_confirmed'])
dfConf.head()
```

```python
data = [go.Scatter(
        x= dfConf.Category,
        y= dfConf.Bahia,
        name= "Confirmados",
        line= dict(color= '#17BECF'),
        opacity= 0.8 )]
py.iplot(data)
```
