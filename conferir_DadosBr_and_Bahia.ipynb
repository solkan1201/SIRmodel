```python
pmtros = {
    'PREDICT_RANGE_def': 150,
    'START_DATE_def': "1/22/20",
    'COUNTRY_CSV_def': ['Brazil'],
    'csv_confirmed': 'tables/time_series_covid19_confirmed_global.csv',
    'csv_death': 'tables/time_series_covid19_deaths_global.csv',
    'csv_recovered': 'tables/time_series_covid19_recovered_global.csv',
    'csv_BAconfirmed': 'tables/casosConfirmadosAc.csv'
}

lsTables = ['csv_confirmed','csv_death','csv_recovered']
```

```python
def load_recovered(path):
        
    df = pd.read_csv(path)
    print("show 5 files of " + 'Brazil')

    country_df = df[df['Country/Region'] == 'Brazil']
    print(country_df.head())      
    
    serie = country_df.iloc[0].loc[pmtros['START_DATE_def']:]  
    #print(serie)    

    return serie, country_df.columns
```

```python
datos = {}
colunas =[]
for nmKeys in lsTables:
    sSerie, colunas = load_recovered(pmtros[nmKeys])
    datos[nmKeys] = [kk for kk in sSerie]

tam = len(datos['csv_death'])
# print("tamanho do dict ", tam)
colunas = [kk for kk in colunas]
colunas = colunas[(len(colunas) - tam):]
datos['date'] = colunas
df = pd.DataFrame(datos)

# print(colunas)
```

```python
df = df.sort_index(axis=0 ,ascending=True)
df.head()
```

```python
def corregirData(dat):
    dats = dat.split('/')
    dia = int(dats[1])
    mes = int(dats[0])
    ano = int('20' + str(dats[2]))
    
    dataj = date(ano,mes,dia)
    return dataj.isoformat()

df['date'] = df['date'].apply(corregirData)
# mdata = corregirData('1/22/20')
# mdata
```

```python
dfBA = pd.read_csv(pmtros['csv_BA'])
dfBA
```

```python
def formatData(dat):    
    dats = dat.split('/')   
    return "{}-{}-{}".format(dats[2],dats[1],dats[0])

dfBA['Category'] = dfBA['Category'].apply(formatData)
# mdata = formatData('06/03/2020')
# mdata
```

```python
dfBA
```

```python
dfBA['Category'][0]
```

```python
df[df['date'] == dfBA['Category'][0]]
```

```python
for cc, data in enumerate(dfBA['Category']):
#     df[df['date'] == data]
    confBA = dfBA[dfBA['Category'] == data]['Confirmados']
    confBA = confBA.values[0]
    
    obitBA = dfBA[dfBA['Category'] == data]['Óbitos']
    obitBA = obitBA.values[0]
    
    df.loc[df['date'] == data, 'confirmed_BA'] = int(confBA)
    df.loc[df['date'] == data, 'obitos_BA'] = int(obitBA)
    
```

```python
df = df.fillna(0)
```

```python
def changeFloatToInt(x):
    return int(x)
df['confirmed_BA'] = df['confirmed_BA'].apply(changeFloatToInt)
df['obitos_BA'] = df['obitos_BA'].apply(changeFloatToInt)
```
